### Import Library & Connect to Google Spreadsheet

```python
import pandas as pd
import plotly.express as px

from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python

spreadsheet = gc.open('IDM4.0 Lahad Datu Feedback Form (for teacher) (Responses)')
# spreadsheet = gc.open_by_key(SHEET_ID)

# worksheet = spreadsheet.sheet1
# worksheet = spreadsheet.get_worksheet(0)
worksheet = spreadsheet.worksheet("(test) 2. Cleaned")

data = worksheet.get_all_values()
df = pd.DataFrame.from_records(data)

df.columns = df.iloc[0] # Set the first row as the header
df = df.drop(0) # Drop the first row as it is now the header
df = df.set_index(df.columns[0]) # Set the first column as the index

df = df.dropna()
df.head(5)
```

### Question - Do you have the resources required and the necessary support to apply what you have learnt?

#### Total Responses: 19 | Mean: 4.79

The result shows highly positive perception of the workshop's overall quality among participants, with a mean rating of 4.79 out of 5 and a median of 5. The majority of responses fall within the upper range of ratings (from 4.5 to 5), suggesting a strong consensus on the workshop's excellence.

```python
fig = px.box(df,
             x="Do you have the resources required and the necessary support to apply what you have learnt?",
             points="all")
fig.update_traces(boxmean=True)


fig.update_layout(
    # xaxis_title='Category',
    # yaxis_title='Value',
    showlegend= False,
    xaxis=dict(
        range=[-1, 6],  # Set the range of the x-axis
        tickmode='array',  # Set tick mode to 'array' to specify custom ticks
    ),
    title=dict(
        text='Titlexxx',
        xanchor='center',
        yanchor= 'top',
        x=0.5,
        y=0.95,
        font=dict(
            family='Montserrat',
            size=16,
            color='black'
        )
    )
)


fig.show()
```
